```python
from pymongo import MongoClient

connection_string='mongodb+srv://REDACTED.waxdm.mongodb.net/?retryWrites=true&w=majority&appName=theCluster'
client = MongoClient(connection_string)
db = client['sample_mflix']
collection = db['movies']
print(collection.find_one())
```

```python
# Exercise 1: Basic Searching & Filtering
# 1. write a query to find the first movie that has the genre "Action"
print("1. Action Movie: ")
print(collection.find_one({"genres": "Action"}))

# 2. write a query to find all movies released after the year 2000
print("\n2. Movies After 2000:")
new_movies = collection.find({"year":{"$gt": 2000}}).limit(5)

for movie in new_movies:
    print(movie)

# 3. write a query to find all movies where the IMDb rating is greater than 8.5
print("\n3. High Rating Movies:")
high_rating = collection.find({"imdb.rating":{"$gt": 8.5}}).limit(5)
for movie in high_rating:
    print(movie)

# 4. write a query to find all movies where the genre contains both "Action" and "Adventure"
print("\n4. Action and Adventure Movies:")
action_adventure = collection.find({"genres":{"$all": ["Action", "Adventure"]}}).limit(5)
for movie in action_adventure:
    print(movie)
```

```python
# Exercise 2: Sorting Results
# 1. write a query to find all movies where the genre is "Comedy" and sort them by IMDb rating in desending order
print("1. Top Comedy Movies:")
top_comedy = collection.find({"genres": "Comedy"}).sort("imdb.rating", -1).limit(5)
for movie in top_comedy:
    print(movie)
# 2. write a query to find all movies where the genre is "Drama" and sort them by release year in ascending order
print("\n2. Worst Drama Movies: ")
worst_drama = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)
for movie in worst_drama:
    print(movie)
```

```python
# Exercise 3: Aggregation Pipeline
# 1. write an aggregation pipeline that calculates the average IMDb rating for each genre
print("Average Ratings by Genre:")
avg_ratings_genre = collection.aggregate([
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit":5}
])
for genre in avg_ratings_genre:
    print(genre)

# 2. write an aggregation pipeline to find the top 5 directors by the average IMDb rating of their movies
print("\nTop 5 Directors:")
top_directors = collection.aggregate([
    {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
])
for director in top_directors:
    print(director)

#3. calculate total number of movies release in each year
print("\nTotal Movies by Year:")
movies_by_year = collection.aggregate([
    {"$group":{"_id":"$year","total_movies":{"$sum":1}}},
    {"$sort":{"_id":1}},
])
for year in movies_by_year:
    print(year)
```

```python
# Exercise 4: Updating and Deleting Documents
# 1. write a query to update the IMDb rating of a movie with the title "The God Father" to 9.5
print("God Father Rating Update:")
gf = collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})
print(gf)

# 2. write a query to update all movies where the genre is "Horror" and set their IMDb rating to 6.0 if it is currently null
print("\nHorror Rating Update:")
horror = collection.update_many({"genres": "Horror", "imdb.rating":{"$exists":False}},{"$set": {"imdb.rating": 6.0}})
all_horror = collection.find({"genres": "Horror"}).limit(5)
for movie in all_horror:
    print(movie)

# 3. write a query to delete all movies that were rleased before the year 1950
print("\nDelete Movies Before 1950:")
before_1950 = collection.delete_many({"year":{"$lt": 1950}})
all_movies = collection.find().limit(5)
for movie in all_movies:
    print(movie)
```

```python
# Exercise 5: Text Search

# 1. ensure the title field is indexed for text search in MongoDB and write a query to search for movies that containt the word "love" in their title
collection.create_index([("title", "text")])
love_movies = collection.find({"$text": {"$search": "love"}}).limit(5)
print("1. Love Movies:")
for movie in love_movies:
    print(movie)

# 2. write a text search query to find movies where the word "war" appears in the title or plot, sorted by IMDb rating (Return only the top 5 results)
print("\n2. Top War Movies:")
war_movies = collection.find({"$text": {"$search": "war"}}).sort("imdb.rating", -1).limit(5)
for movie in war_movies:
    print(movie)
```
